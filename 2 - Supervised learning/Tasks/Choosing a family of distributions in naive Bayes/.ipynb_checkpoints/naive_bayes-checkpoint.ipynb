Все, что нужно будет делать — запускать методы из sklearn. Вам предлагается выяснить, какое распределение лучше использовать в наивном байесовском классификаторе в зависимости от вида признаков.

Загрузите датасеты digits и breast_cancer из sklearn.datasets. Выведите несколько строчек из обучающих выборок и посмотрите на признаки. С помощью sklearn.cross_validation.cross_val_score c настройками по умолчанию и вызова метода mean() у возвращаемого этой функцией numpy.ndarray, сравните качество работы наивных байесовских классификаторов на этих двух датасетах. Для сравнения предлагается использовать BernoulliNB, MultinomialNB и GaussianNB. Насколько полученные результаты согласуются с рекомендациями из лекций?

Два датасета, конечно, еще не повод делать далеко идущие выводы, но при желании вы можете продолжить исследование на других выборках (например, из UCI репозитория).

```python
from sklearn import datasets, naive_bayes
from sklearn.model_selection import cross_val_score
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()

digits_X = digits.data
digits_y = digits.target
breast_cancer_X = breast_cancer.data
breast_cancer_y = breast_cancer.target
```

```python
digits_X[:5]
```

```python
breast_cancer_X[:5]
```

```python
bernb = naive_bayes.BernoulliNB()

dscore_1 = cross_val_score(bernb, digits_X, digits_y).mean()
print('Cross_val_score of BernoulliNB for digits: {}\n'.format(dscore_1))

bcscore_1 = cross_val_score(bernb, breast_cancer_X, breast_cancer_y).mean()
print('Cross_val_score of BernoulliNB for breast cancer: {}'.format(bcscore_1))
```

```python
multinb = naive_bayes.MultinomialNB()

dscore_2 = cross_val_score(multinb, digits_X, digits_y).mean()
print('Cross_val_score of MultinomialNB for digits: {}\n'.format(dscore_2))

bcscore_2 = cross_val_score(multinb, breast_cancer_X, breast_cancer_y).mean()
print('Cross_val_score of MultinomialNB for breast cancer: {}'.format(bcscore_2))
```

```python
gaussnb = naive_bayes.GaussianNB()

dscore_3 = cross_val_score(gaussnb, digits_X, digits_y).mean()
print('Cross_val_score of GaussianNB for digits: {}\n'.format(dscore_3))

bcscore_3 = cross_val_score(gaussnb, breast_cancer_X, breast_cancer_y).mean()
print('Cross_val_score of GaussianNB for breast cancer: {}'.format(bcscore_3))
```
